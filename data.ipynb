# Creat a little Dataset for first test and gain time

```python
df_train = pd.read_csv('./data/fashion-mnist_train.csv')
```

```python
label_tain = df_train['label']
data_train = df_train.drop('label', axis=1)
```

```python
def setupDatasets(ratio: int):
    global label_tain, data_train
    label_tain = label_tain[:len(label_tain)//ratio]
    data_train = data_train[:len(data_train)//ratio]
    return data_train, label_tain
```

```python
data_train, label_tain = setupDatasets(ratio=10)
```
